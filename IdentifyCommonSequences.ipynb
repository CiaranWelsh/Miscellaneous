# Longest common subsequence
The code in the above cell was directly copied from [stackoverflow](https://stackoverflow.com/questions/2892931/longest-common-substring-from-more-than-two-strings-python).
```python
def long_substr(data):
    substr_list=[]
    substr = ''
    if len(data) > 1 and len(data[0]) > 0:
        for i in range(len(data[0])):
            for j in range(len(data[0])-i+1):
                if j > len(substr) and is_substr(data[0][i:i+j], data):
                    substr_list.append(data[0][i:i+j])
                    substr = data[0][i:i+j]
    return substr_list

def is_substr(find, data):
    if len(data) < 1 and len(find) < 1:
        return False
    for i in range(len(data)):
        if find not in data[i]:
            return False
    return True

## remove the newline characters from the sequences
id1 = id1.replace('\n','')
smad7 = smad7.replace('\n', '')
rara = rara.replace('\n', '')
vegfa = vegfa.replace('\n', '')
junb = junb.replace('\n', '')

## test the function 
test = long_substr(['I am typing a sentence',
                    ' sentence is not',
                    'making sence'])  
print('This is a test for function: {}'.format(test))
'''
This is the smad3 binding site:
    CAGACA
'''


common_subsequences = (long_substr([id1, smad7, rara, vegfa, junb]))

```
